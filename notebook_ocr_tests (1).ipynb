```python
# ---------------------------------------------------------
# OCR TEST SETUP
# ---------------------------------------------------------
# This cell initializes the src.ocr module so we can import
# read_pdf_text() and read_image_text() without errors.
# ---------------------------------------------------------

import os, sys, types, re

def _load_fallback_shim():
    """Creates an in-memory src package with minimal OCR + extractor functionality."""
    print("Loading fallback shim for src modules...")

    # Try OCR libraries, but don't fail if missing
    try:
        import pytesseract
        from PIL import Image
        import pdfplumber
    except Exception:
        pytesseract = None
        Image = None
        pdfplumber = None

    # ---- OCR: Image ----
    def read_image_text(path):
        if pytesseract is None or Image is None:
            return "[OCR not available: pytesseract or PIL missing]"
        try:
            img = Image.open(path).convert("RGB")
            return pytesseract.image_to_string(img)
        except Exception:
            return "[Failed to read image file]"

    # ---- OCR: PDF ----
    def read_pdf_text(path):
        if pdfplumber is None:
            return "[PDF extraction not available: pdfplumber missing]"
        try:
            pages = []
            with pdfplumber.open(path) as pdf:
                for page in pdf.pages:
                    txt = page.extract_text()
                    if txt:
                        pages.append(txt)
            return "\n".join(pages) if pages else "[No text found in PDF]"
        except Exception:
            return "[Failed to read PDF file]"

    # ---- Create src.* modules ----
    src_mod = types.ModuleType("src")
    ocr_mod = types.ModuleType("src.ocr")

    ocr_mod.read_image_text = read_image_text
    ocr_mod.read_pdf_text = read_pdf_text

    src_mod.ocr = ocr_mod

    sys.modules["src"] = src_mod
    sys.modules["src.ocr"] = ocr_mod
    print("Fallback src.ocr loaded successfully.")

# ---------------------------------------------------------
# Try to run the real project file if it exists
# ---------------------------------------------------------

if os.path.exists("IntelligentFormAgent.py.ipy"):
    print("Running IntelligentFormAgent.py to load src modules...")
    try:
        get_ipython().run_line_magic("run", "IntelligentFormAgent.py.ipy")
    except Exception as e:
        print("Error running IntelligentFormAgent.py:", e)
        print("Switching to fallback shim.")
        _load_fallback_shim()
else:
    print("IntelligentFormAgent.py not found — using fallback shim.")
    _load_fallback_shim()



# ---------------------------------------------------------
# OCR TEST EXECUTION
# ---------------------------------------------------------

from src.ocr import read_pdf_text, read_image_text

pdf_file = "form1_autonomize_ai.pdf"
img_file = "ChatGPT Image Nov 17, 2025, 07_39_22 PM.png"

print("\n================ PDF OCR TEST ================")
print(read_pdf_text(pdf_file))

print("\n================ IMAGE OCR TEST ================")
print(read_image_text(img_file))
```
